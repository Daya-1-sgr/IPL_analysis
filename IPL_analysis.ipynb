```python
df_first_over_data=df[df['over']==0]
df_first_over_data.tail(10)
```

```python
batting_stats=df_first_over_data.groupby('batter').agg({'ball':'count','batsman_runs':'sum','is_wicket':'sum'}).reset_index()
batting_stats.head()
```

```python
batting_stats['strike_rate']=(batting_stats['batsman_runs']/batting_stats['ball'])*100
batting_stats['batting_average']=batting_stats['batsman_runs']/batting_stats['is_wicket']
batting_stats[batting_stats['batsman_runs']>=50].sort_values(by='strike_rate',ascending=False).head()
```

```python
finisher_stats=df[df['over']>13]
finisher_stats.head()
```

```python
finishing_stats=finisher_stats.groupby('batter').agg({'ball':'count','batsman_runs':'sum','is_wicket':'sum'}).reset_index()
finishing_stats.head()
```

```python
finishing_stats['strike_rate']=(finishing_stats['batsman_runs']/finishing_stats['ball'])*100
finishing_stats['batting_average']=finishing_stats['batsman_runs']/finishing_stats['is_wicket']
finishing_stats[finishing_stats['batsman_runs']>=50].sort_values(by='strike_rate',ascending=False).head(20)
```
